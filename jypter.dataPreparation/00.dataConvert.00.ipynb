```python
tabelDic = {'tr':[], 'aa_length':[], 'protein':[], 'OS':[], 'OX':[], 'GN':[], 'PE':[], 'SV':[] }
for ii in range(len(tempTableDf)):
    # if ii == 10000: break   
    rowDf = tempTableDf.iloc[ii]
    infoStr, aaLen = rowDf['info'][0:-1] , rowDf['aa_length']
    if ii % 200000 == 0:
        print(ii, infoStr)
    # print(ii, "-->", infoStr, aaLen)
    trStr = infoStr.split(' ')[0]
    trStr = trStr[4:len(trStr)+1]
    
    proteinName = infoStr.split('OS=')[0].split(' ')[1:-1]
    proteinName = ' '.join(proteinName)
    
    OSname = infoStr.split('OS=')[1].split('OX=')[0]
    OXname = infoStr.split('OX=')[1].split(' ')[0]
    # print(ii, infoStr)
    GNname = infoStr.split('GN=')[1].split('PE=')[0] if 'GN=' in infoStr else np.nan
    PEname = infoStr.split('PE=')[1].split('SV=')[0]
    SVname = infoStr.split('SV=')[1]
    
    tabelDic['tr'].append(trStr)
    tabelDic['aa_length'].append(aaLen)
    tabelDic['protein'].append(proteinName)
    tabelDic['OS'].append(OSname)
    tabelDic['OX'].append(OXname)
    tabelDic['GN'].append(GNname)
    tabelDic['PE'].append(PEname)
    tabelDic['SV'].append(SVname)
    
tableDf = pd.DataFrame(data=tabelDic)
tableDf.to_csv('./data/tableDf.csv', index=False)
tableDf = pd.read_csv('./data/tableDf.csv')
tableDf.tail(n=10)
```
