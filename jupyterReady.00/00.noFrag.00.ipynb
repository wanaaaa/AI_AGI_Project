```python
homoDf = noFragTableDf[noFragTableDf['OS'].str.contains('Homo sapiens')]
homoLenLt = list(homoDf['aa_length'])

homoLen10Dic = {}
homoLenNo10Lt = [x - (x % 10) for x in homoLenLt]
for ii, cLen in enumerate(homoLenNo10Lt):
    # if ii == 1: break
    if cLen in homoLen10Dic.keys():
        homoLen10Dic[cLen] += 1
    else:
        homoLen10Dic[cLen] = 1
print('-----------------')
```

```python
xAAlen = list(homoLen10Dic.keys())[0:100]
xAAlen.sort()
yNum = [homoLen10Dic[x] for x in xAAlen]

xAver = xAAlen[0:len(xAAlen) -5]
yAver = [mean(yNum[x:x+5]) for x in range(len(xAver))]
print(len(xAAlen), xAAlen)
print(yNum)
print(homoLen10Dic)


# numberOfRow, numberOfCol = 1 , 1
# fig, axes = plt.subplots(numberOfRow, numberOfCol, figsize=(numberOfCol*5,numberOfRow*3) )
# plt.subplots_adjust(
#     # left=0.1,
#     # bottom=0.1, 
#     # right=0.9, 
#     # top=0.9, 
#     wspace=0.3, 
#     hspace=0.4
# )

# wPlot = sns.lineplot(x=xAver, y=yAver, ax=axes, color='blue')
# wPlot = sns.barplot(x=xAAlen, y=yNum, ax=axes, color='red' )
```
